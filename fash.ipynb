```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
import tensorflow as tf
from keras.models import Model
from keras.layers import Input, Conv2D, MaxPooling2D, UpSampling2D, Flatten, Dense, Reshape
from keras.optimizers import Adam
from keras.callbacks import ModelCheckpoint, EarlyStopping
from sklearn.model_selection import train_test_split
import cv2

class FashionEncoder:
    def __init__(self, img_height=128, img_width=128, channels=3):
        self.img_height = img_height
        self.img_width = img_width
        self.channels = channels
        self.encoding_dim = 128

    def load_and_preprocess_images(self, image_dir="/content/down/downloaded_images"):
        """Load and preprocess images for training"""
        images = []
        image_paths = []

        print("Loading images...")

        for root, dirs, files in os.walk(image_dir):
            for file in files:
                if file.lower().endswith(('.jpg', '.jpeg', '.png', '.gif', '.bmp')):
                    filepath = os.path.join(root, file)
                    try:
                        # Load and preprocess image
                        img = Image.open(filepath).convert('RGB')
                        img = img.resize((self.img_width, self.img_height))
                        img_array = np.array(img) / 255.0  # Normalize to [0, 1]

                        images.append(img_array)
                        image_paths.append(filepath)

                        if len(images) % 100 == 0:
                            print(f"Loaded {len(images)} images...")

                    except Exception as e:
                        print(f"Error loading {filepath}: {e}")
                        continue

        print(f"Total images loaded: {len(images)}")
        return np.array(images), image_paths

    def build_autoencoder(self):
        """Build the autoencoder model for feature learning"""
        # Encoder
        input_img = Input(shape=(self.img_height, self.img_width, self.channels))

        # Encoder layers
        x = Conv2D(32, (3, 3), activation='relu', padding='same')(input_img)
        x = MaxPooling2D((2, 2), padding='same')(x)
        x = Conv2D(64, (3, 3), activation='relu', padding='same')(x)
        x = MaxPooling2D((2, 2), padding='same')(x)
        x = Conv2D(128, (3, 3), activation='relu', padding='same')(x)
        x = MaxPooling2D((2, 2), padding='same')(x)
        x = Conv2D(256, (3, 3), activation='relu', padding='same')(x)

        # Flatten for dense layers
        x = Flatten()(x)
        encoded = Dense(self.encoding_dim, activation='relu')(x)

        # Decoder
        x = Dense(256 * 16 * 16, activation='relu')(encoded)
        x = Reshape((16, 16, 256))(x)
        x = Conv2D(256, (3, 3), activation='relu', padding='same')(x)
        x = UpSampling2D((2, 2))(x)
        x = Conv2D(128, (3, 3), activation='relu', padding='same')(x)
        x = UpSampling2D((2, 2))(x)
        x = Conv2D(64, (3, 3), activation='relu', padding='same')(x)
        x = UpSampling2D((2, 2))(x)
        x = Conv2D(32, (3, 3), activation='relu', padding='same')(x)
        decoded = Conv2D(self.channels, (3, 3), activation='sigmoid', padding='same')(x)

        # Create models
        autoencoder = Model(input_img, decoded)
        encoder = Model(input_img, encoded)

        return autoencoder, encoder

    def train_model(self, images, epochs=50, batch_size=32, validation_split=0.2):
        """Train the autoencoder model"""
        # Build the model
        autoencoder, encoder = self.build_autoencoder()

        # Compile the model
        autoencoder.compile(optimizer=Adam(learning_rate=0.001), loss='mse')

        # Print model summary
        print("Autoencoder Architecture:")
        autoencoder.summary()

        # Callbacks
        callbacks = [
            ModelCheckpoint('fashion_encoder_best.h5', save_best_only=True, monitor='val_loss'),
            EarlyStopping(patience=10, restore_best_weights=True)
        ]

        # Train the model
        history = autoencoder.fit(
            images, images,  # Autoencoder trains on input=output
            epochs=epochs,
            batch_size=batch_size,
            validation_split=validation_split,
            callbacks=callbacks,
            verbose=1
        )

        # Save the final models
        autoencoder.save('fashion_autoencoder.h5')
        encoder.save('fashion_encoder.h5')

        return autoencoder, encoder, history

    def plot_training_history(self, history):
        """Plot training and validation loss"""
        plt.figure(figsize=(12, 4))

        plt.subplot(1, 2, 1)
        plt.plot(history.history['loss'], label='Training Loss')
        plt.plot(history.history['val_loss'], label='Validation Loss')
        plt.title('Model Loss')
        plt.xlabel('Epoch')
        plt.ylabel('Loss')
        plt.legend()

        plt.tight_layout()
        plt.savefig('training_history.png')
        plt.show()

    def test_reconstruction(self, autoencoder, test_images, num_images=5):
        """Test the autoencoder reconstruction quality"""
        # Get reconstructions
        reconstructed = autoencoder.predict(test_images[:num_images])

        # Plot original vs reconstructed
        fig, axes = plt.subplots(2, num_images, figsize=(15, 6))

        for i in range(num_images):
            # Original
            axes[0, i].imshow(test_images[i])
            axes[0, i].set_title('Original')
            axes[0, i].axis('off')

            # Reconstructed
            axes[1, i].imshow(reconstructed[i])
            axes[1, i].set_title('Reconstructed')
            axes[1, i].axis('off')

        plt.tight_layout()
        plt.savefig('reconstruction_comparison.png')
        plt.show()

    def extract_features(self, encoder, images):
        """Extract features from images using trained encoder"""
        features = encoder.predict(images)
        return features

    def save_features_to_csv(self, features, image_paths, filename="image_features.csv"):
        """Save extracted features to CSV file"""
        # Create DataFrame with features
        feature_columns = [f'feature_{i}' for i in range(features.shape[1])]
        df = pd.DataFrame(features, columns=feature_columns)
        df['image_path'] = image_paths

        # Save to CSV
        df.to_csv(filename, index=False)
        print(f"Features saved to {filename}")
        return df

def main():
    # Initialize the encoder
    encoder_trainer = FashionEncoder()

    # Load and preprocess images
    images, image_paths = encoder_trainer.load_and_preprocess_images()

    if len(images) == 0:
        print("No images found. Please run the image download script first.")
        return

    # Train the model
    print("Training autoencoder...")
    autoencoder, encoder, history = encoder_trainer.train_model(images)

    # Plot training history
    encoder_trainer.plot_training_history(history)

    # Test reconstruction quality
    encoder_trainer.test_reconstruction(autoencoder, images)

    # Extract features from all images
    print("Extracting features...")
    features = encoder_trainer.extract_features(encoder, images)

    # Save features to CSV
    feature_df = encoder_trainer.save_features_to_csv(features, image_paths)

    print(f"Feature extraction complete!")
    print(f"Feature shape: {features.shape}")
    print(f"Features saved to image_features.csv")

if __name__ == "__main__":
    main()
```

```python
import pandas as pd
import numpy as np
import tensorflow as tf
import matplotlib.image as mpimg
from skimage.transform import resize
from keras.models import Sequential
from keras.layers import Dense
from sklearn.preprocessing import MinMaxScaler
import os

os.environ['TF_CPP_MIN_LOG_LEVEL'] = '2'

# ------------------- Load Encoder -------------------
encoder_path = "/content/encoder.keras"
encoder = tf.keras.models.load_model(encoder_path, compile=False)

# ------------------- Load Dataset -------------------
df = pd.read_csv("/content/real_fashion_data.csv")

# ------------------- Clean Ratings and Reviews -------------------
df["rating"] = pd.to_numeric(df["rating"], errors='coerce').fillna(3.0)  # neutral rating
if "no_of_reviews" not in df.columns:
    df["no_of_reviews"] = 10  # default review count
df["no_of_reviews"] = pd.to_numeric(df["no_of_reviews"], errors='coerce').fillna(10).astype(int)

# ------------------- Load and Resize Images -------------------
images = []
valid_indices = []

for i in range(len(df)):
    img_path = os.path.join("/content/crop/cropped_tshirts", f"tshirt_{i}_0.jpg")
    if os.path.exists(img_path):
        try:
            img = mpimg.imread(img_path)
            img = resize(img, (320, 192, 3))
            images.append(img)
            valid_indices.append(i)
        except Exception as e:
            print(f"❌ Error reading {img_path}: {e}")

if len(images) == 0:
    raise ValueError("❌ No valid images found in /content/crop/cropped_tshirts")

print(f"✅ Loaded {len(images)} valid images.")

# ------------------- Match DataFrame with Valid Images -------------------
df = df.loc[valid_indices].reset_index(drop=True)

# ------------------- Encode Images -------------------
image_array = np.array(images)
encodings = encoder.predict(image_array)
df["encodings"] = [e.flatten() for e in encodings]

# ------------------- Compute Popularity -------------------
def pop_met(n, s):
    n = np.array(n)
    s = np.array(s)
    top = s * (15 + n)
    bott = n + 5 * s + 1e-5  # epsilon to avoid divide-by-zero
    return top / bott

df["popularity"] = pop_met(df["no_of_reviews"], df["rating"])

scaler = MinMaxScaler()
df["popularity"] = scaler.fit_transform(df["popularity"].values.reshape(-1, 1))

df["popularity"].replace([np.inf, -np.inf], np.nan, inplace=True)
df = df.dropna(subset=["popularity", "encodings"])
df = df[df["encodings"].apply(lambda x: np.isfinite(x).all())]

X = np.array(df["encodings"].tolist())
y = np.array(df["popularity"]).reshape(-1, 1)


model = Sequential([
    Dense(1920, input_dim=1920, activation='relu'),
    Dense(1024, activation='relu'),
    Dense(512, activation='relu'),
    Dense(128, activation='relu'),
    Dense(1)
])
model.compile(loss='mse', optimizer='adam', metrics=['mae', 'mse'])


model.fit(X, y, batch_size=32, epochs=50, validation_split=0.2)


model.save("pm_model_components/pm_model.keras")
df_sorted = df.sort_values(by='popularity', ascending=False)
df_sorted.to_csv("df_sorted.csv", index=False)

print("✅ Model training complete.")
print("✅ Sorted DataFrame saved as df_sorted.csv.")
print("✅ PM model saved to pm_model_components/pm_model.keras.")
```
